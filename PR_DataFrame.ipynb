## Soal 2

```python
arr1 = np.arange(15)
pd.Series(data=arr1,index=arr1)
```

```python
col0 = pd.Series(arr1[(arr1%2==0) & (arr1 <=10)])
col1 = pd.Series(arr1[(arr1%2!=0) & (arr1 <=11)])
col2 = pd.Series(arr1[(arr1%2==0) & (arr1 >=2) & (arr1 <= 12)])
col3 = pd.Series(arr1[(arr1%2!=0) & (arr1 >=3) & (arr1 <= 13)])
k = {'Column 0' : col0,'Column 1' : col1,'Column 2' : col2, 'Column 3': col3}
d = pd.DataFrame(k)
d
```

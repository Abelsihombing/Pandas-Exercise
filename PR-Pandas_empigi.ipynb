# Soal 3

```python
mpg['origin'].unique()
```

```python
# If the horsepower is lower and equal to 105, then it is lower. if it is higher than 105 then it is upper
# cara cornel
def add_horse_origin(x):
    if x['horsepower'] <= 105:
        return 'lower {}'.format(x['origin'])
    else:
        return 'upper {}'.format(x['origin'])
```

```python
mpg['horse_origin'] = mpg[['horsepower','origin']].apply(add_horse_origin, axis= 1)
```

```python
mpg['horse_origin'].value_counts()
```

```python
# Disini dibuat aggregate dr statistik dasar yang telah digroupby
#berdasrkan origin
mpg_origin = mpg.groupby('origin').agg(['mean','std','sum','count','min','max'])
mpg_origin['horsepower'].reset_index()
```

```python
# Disini dibuat merge event utk mpg dan aggregate khusus origin saja
group_agg = mpg_origin['horsepower'].copy()
pd.merge(mpg, group_agg.reset_index(), on = 'origin')
```
